# Image Segmentation

Image Segmentation is one of the most important tasks in Computer Vision. Unlike image classification (which predicts one label for the whole image) or object detection (which draws bounding boxes around objects), image segmentation predicts a label for **every single pixel** in the image.

This makes segmentation a **dense prediction problem**.

# Introduction to Image Segmentation

## What is Image Segmentation?

Image segmentation is the process of dividing an image into meaningful regions by assigning each pixel a class label.

For example:

* In a medical image → separate tumor from healthy tissue
* In self-driving cars → separate road, pedestrians, cars, sky
* In agriculture → separate crops from weeds

Instead of predicting:

* "There is a cat"

Segmentation predicts:

* "These exact pixels belong to the cat"

---

## Why is it Important?

Segmentation provides:

* Precise object boundaries
* Spatial understanding
* Scene structure information
* Better decision-making for AI systems

It enables machines to "understand" scenes at a much deeper level.

---

# Types of Image Segmentation

## Semantic Segmentation

Each pixel is assigned a class label.

Example:

* All cars → same label
* All people → same label

It does NOT differentiate between different instances of the same class.

**Use Cases:**
- Road scene understanding (autonomous vehicles)
- Satellite imagery (land cover classification)

## Instance Segmentation

Each object instance is separated.

Example:

* Car 1 → different mask
* Car 2 → different mask

This combines object detection + segmentation.

**Use Cases:**
- Counting objects (e.g., cells in microscopy)
- Retail analytics (counting people or products)

## Panoptic Segmentation

Combines:

* Semantic segmentation (background stuff)
* Instance segmentation (countable objects)

**Use Cases:**
- Urban scene parsing
- Robotics navigation

| Type                | Differentiates Instances? | Example Use Case                |
|---------------------|--------------------------|---------------------------------|
| Semantic            | No                       | Land cover mapping              |
| Instance            | Yes                      | Counting people in a crowd      |
| Panoptic            | Yes (for objects)        | Autonomous driving              |

# Downsampling

## What is Downsampling?

Downsampling reduces the spatial resolution of feature maps.

Example:

* 256x256 → 128x128 → 64x64

## Why Do We Use It?

1. Reduce computation
2. Increase receptive field (see more of the image)
3. Extract high-level features
4. Capture global context

## Methods of Downsampling

### Max Pooling

* Takes maximum value from window
* Keeps strong features
* Can lose fine details

### Average Pooling

* Takes average of window
* Smoother, but may blur features

### Strided Convolution

* Convolution with stride > 1
* Learnable downsampling
* Can combine feature extraction and downsampling

## Role in Segmentation

Downsampling:

* Helps extract semantic meaning
* But reduces spatial details
* Needs upsampling later to restore resolution

**Tip:** Too much downsampling can make it hard to recover fine boundaries in segmentation.

---

# Upsampling / Transposed Convolution

## Why Upsampling?

Segmentation requires output size = input size.

After downsampling, we must recover spatial resolution.

## Upsampling Methods

### Nearest Neighbor Interpolation

Simple pixel duplication. Fast, but can look blocky.

### Bilinear Interpolation

Smooth interpolation. Better visual quality, but not learnable.

### Transposed Convolution (Deconvolution)

Learnable upsampling layer. Can adapt to data, but may introduce artifacts.

### Pixel Shuffle

Rearranges channels to spatial resolution. Used in super-resolution.

---

## Transposed Convolution Details

* Uses learnable kernels
* Can introduce checkerboard artifacts (watch for this in outputs)
* Important in decoder architectures
* Often combined with skip connections to recover details

# Segmentation Loss Functions

Loss functions measure how close predicted mask is to ground truth.

## Cross Entropy Loss

Used for multi-class segmentation.

* Penalizes incorrect class predictions for each pixel
* Sensitive to class imbalance

## Binary Cross Entropy

Used for binary segmentation.

* Each pixel is treated as a binary classification

## Dice Loss

Measures overlap between prediction and ground truth.

Dice Coefficient:

Dice = 2 * (Intersection) / (Prediction + Ground Truth)

Good for medical segmentation and imbalanced data.

## IoU Loss (Jaccard Loss)

IoU = Intersection / Union

* Similar to Dice, but penalizes false positives more

## Focal Loss

Handles class imbalance by focusing on hard-to-classify pixels.

## Combined Losses

Often combine:

* Cross entropy + Dice
* Helps balance pixel-wise accuracy and overlap

**Tip:** Always check class balance in your dataset before choosing a loss function.

# Fully Convolutional Networks (FCNs)

## What are FCNs?

FCNs are CNNs without fully connected layers.

They replace dense layers with convolutional layers, allowing input images of any size.

## Key Idea

Input: Any image size
Output: Corresponding segmentation map

## Architecture

* Encoder (downsampling)
* Decoder (upsampling)

## Skip Connections

Combine:

* Low-level features (edges)
* High-level features (semantics)

## Importance

FCNs were the first deep learning architecture for semantic segmentation.

**Limitations:**
- Coarse outputs (low resolution)
- Struggles with fine boundaries

# U-Net

## What is U-Net?

U-Net is a segmentation architecture widely used in medical imaging.

It has a U-shaped architecture with symmetric encoder and decoder paths.

## Structure

### Encoder (Contracting Path)

* Conv
* ReLU
* Pooling

### Decoder (Expanding Path)

* Transposed convolution
* Concatenation
* Conv layers

## Skip Connections

Key idea:

* Copy feature maps from encoder
* Concatenate with decoder

This preserves fine details and helps recover spatial information lost during downsampling.

## Why U-Net is Powerful

* Works with small datasets
* Preserves localization
* Simple and efficient
* Widely used in biomedical image segmentation

**Limitations:**
- May not scale well to very large images
- Sensitive to hyperparameters

# Mask R-CNN

## What is Mask R-CNN?

Mask R-CNN extends Faster R-CNN by adding a segmentation branch.

It performs:

* Object detection
* Bounding box regression
* Instance segmentation

## Architecture Components

1. Backbone (ResNet, etc.)
2. Region Proposal Network (RPN)
3. ROI Align
4. Classification head
5. Bounding box head
6. Mask head

## ROI Align

Improves spatial accuracy compared to ROI Pooling by avoiding quantization errors.

## Why Mask R-CNN is Important

* Accurate instance segmentation
* Widely used in real-world applications
* Flexible and extendable to other tasks (e.g., keypoint detection)

**Limitations:**
- Computationally intensive
- Requires large labeled datasets

# Evaluation Metrics

## Pixel Accuracy

**Definition:**

Pixel accuracy is the ratio of correctly classified pixels to the total number of pixels in the image. It is a simple and intuitive metric for segmentation tasks.

**Formula:**

Pixel Accuracy = (Number of correctly classified pixels) / (Total number of pixels)

**Pros:**
- Easy to compute and understand
- Useful for balanced datasets

**Cons:**
- Can be misleading for imbalanced datasets (e.g., large background regions)

---

## Mean IoU (mIoU)

**Definition:**

Mean Intersection over Union (mIoU) is the average IoU across all classes. IoU for a class is the ratio of the intersection to the union of the predicted and ground truth masks for that class.

**Formula:**

IoU = (True Positive) / (True Positive + False Positive + False Negative)

mIoU = Average IoU over all classes

**Pros:**
- Robust to class imbalance
- Standard metric for segmentation challenges

**Cons:**
- Can be low for rare classes

---

## Dice Score

**Definition:**

Dice Score (Dice Coefficient) measures the overlap between the predicted and ground truth masks. It is especially popular in medical image segmentation.

**Formula:**

Dice = 2 * (Intersection) / (Prediction + Ground Truth)

**Pros:**
- Sensitive to overlap
- Good for imbalanced datasets

**Cons:**
- Can be less interpretable for multi-class problems

---

## Precision and Recall

**Definition:**

Precision measures the proportion of predicted positive pixels that are actually positive. Recall measures the proportion of actual positive pixels that are correctly predicted.

**Formulas:**

Precision = (True Positive) / (True Positive + False Positive)
Recall = (True Positive) / (True Positive + False Negative)

**Pros:**
- Useful for imbalanced datasets
- Can be combined into F1-score

**Cons:**
- Need to be interpreted together for a full picture
