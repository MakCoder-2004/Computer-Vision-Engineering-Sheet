# Object Detection

Object detection is a computer vision technique that identifies and locates objects within images or videos. It not only classifies objects but also draws bounding boxes around them, enabling applications like autonomous driving, surveillance, and image analysis. Popular approaches include traditional methods (like Haar cascades) and deep learning models (such as YOLO, SSD, and Faster R-CNN), which have significantly improved detection accuracy and speed.

### Input
   - Image

### Output
`[x1, y1, x2, y2, conf_score, class_name]`
   - Bounding Box (Object Location)
   - Confidence Score (From 0 to 1)
   - Object Category (Class Name)

## Object Detection Metrics

### 1. Precision
Measures the proportion of correctly predicted positive detections out of all positive predictions.

- Precision = TP / (TP + FP)

### 2. Recall
Measures the proportion of correctly predicted positive detections out of all actual positives.

- Recall = TP / (TP + FN)

### 3. Intersection over Union (IoU)
Evaluates the overlap between the predicted bounding box and the ground truth box. Higher is better.

- IoU = Area of Overlap / Area of Union

### 4. Average Precision (AP)
Summarizes the precision-recall curve as the weighted mean of precisions at each threshold.

### 5. Mean Average Precision (mAP)
The mean of Average Precision (AP) across all classes, commonly used as the main metric for object detection performance.

### 6. F1 Score
The harmonic mean of precision and recall.

- F1 = 2 × (Precision × Recall) / (Precision + Recall)

### Terminology

- **TP**: True Positives
- **FP**: False Positives
- **FN**: False Negatives

## Bounding Box

### 1. Bounding Box Representation
A bounding box is typically represented using one of the following formats:

**(a) Corner Format (Pascal VOC)**
- (x_min, y_min, x_max, y_max)

Where:
- x_min, y_min → top-left corner
- x_max, y_max → bottom-right corner

Width  = x_max − x_min
Height = y_max − y_min

**(b) Center Format (YOLO)**
(x_center, y_center, width, height)

Where:
- x_center, y_center → center of the box
- width, height → box dimensions

Conversion from corner format:
   - x_center = (x_min + x_max) / 2
   - y_center = (y_min + y_max) / 2
   - width  = x_max − x_min
   - height = y_max − y_min

### 2. Bounding Box Area
- Area = width × height

### 3. Intersection Area
To compute overlap between two boxes:

- x_left   = max(x_min1, x_min2)
- y_top    = max(y_min1, y_min2)
- x_right  = min(x_max1, x_max2)
- y_bottom = min(y_max1, y_max2)

**Intersection Width**  = max(0, x_right − x_left)
**Intersection Height** = max(0, y_bottom − y_top)

- Intersection Area = Intersection Width × Intersection Height

### 4. Union Area
- Union Area = Area_box1 + Area_box2 − Intersection Area

### 5. Aspect Ratio
- Aspect Ratio = width / height

### 6. Bounding Box Regression
Used to refine predicted bounding boxes.

Typical regression targets:
   - tx = (x_gt − x_pred) / width_pred
   - ty = (y_gt − y_pred) / height_pred
   - tw = log(width_gt / width_pred)
   - th = log(height_gt / height_pred)

Where:
- gt   → ground truth
- pred → predicted box

## Object Detection Tools & Libraries

### 1. OpenCV
An open-source computer vision library widely used for image and video processing.

Key Features:
- Image preprocessing (resize, blur, thresholding, etc.)
- Drawing and manipulating bounding boxes
- Real-time video processing
- DNN module for running pre-trained object detection models (YOLO, SSD, etc.)

Common Use Cases:
- Classical computer vision tasks
- Real-time camera applications
- Edge-device deployment


### 2. YOLO (You Only Look Once)
A real-time object detection algorithm that treats detection as a single regression problem.

Example: YOLOv11 (latest versions improve speed and accuracy)

Key Features:
- Single-stage detector
- High inference speed
- End-to-end training
- Supports object detection, segmentation, and pose estimation

Common Use Cases:
- Real-time detection systems
- Autonomous systems
- Surveillance and tracking


### 3. Detectron2
A powerful object detection and segmentation library developed by Facebook AI Research (FAIR).

Key Features:
- Built on PyTorch
- Supports Faster R-CNN, Mask R-CNN, RetinaNet, etc.
- Modular and highly customizable
- State-of-the-art research implementations

Common Use Cases:
- Research and experimentation
- Instance segmentation
- Advanced detection pipelines


### 4. MediaPipe
A cross-platform framework developed by Google for building multimodal ML pipelines.

Key Features:
- Real-time performance
- Pre-built solutions (Face Detection, Pose, Hands, Holistic)
- Lightweight and optimized for mobile devices
- Easy integration with Python, C++, Android, and iOS

Common Use Cases:
- Face tracking
- Gesture recognition
- Human pose estimation
- Mobile and embedded AI applications


### 5. TensorFlow Object Detection API
A framework built on TensorFlow for training and deploying object detection models.

Key Features:
- Pre-trained models (SSD, Faster R-CNN, EfficientDet)
- Custom dataset training
- Easy model exporting

Common Use Cases:
- Custom object detection projects
- Production ML pipelines


### 6. PyTorch + Torchvision
Provides detection models and utilities for object detection.

Key Features:
- Pre-trained models (Faster R-CNN, RetinaNet, SSD)
- Flexible and research-friendly
- Easy customization

Common Use Cases:
- Deep learning research
- Custom detection architectures


### 7. MMDetection
An open-source object detection toolbox based on PyTorch.

Key Features:
- Large collection of SOTA models
- Modular design
- Strong benchmarking support

Common Use Cases:
- Academic research
- Large-scale detection experiments
