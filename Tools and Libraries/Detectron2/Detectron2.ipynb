# Detectron2

Detectron2 is a state-of-the-art open-source library developed by Facebook AI Research (FAIR) for object detection, segmentation, and other computer vision tasks. It is built on PyTorch and provides a flexible, modular design that makes it easy to implement and experiment with cutting-edge deep learning models.
- Detectron2 was introduced in 2019 as the successor to the original Detectron library, offering improved speed, scalability, and extensibility. It supports a wide range of vision tasks, including object detection, instance segmentation, keypoint detection, and panoptic segmentation.

- Detectron2 is widely used in both academic research and industry for building high-performance computer vision systems, thanks to its rich model zoo, strong community support, and production-ready features.
